````
import csv
f = open('인공지능_직업별_현재·부족_인력_20230619112429.csv', encoding='cp949') #encoding='cp949' 가 필요하면 넣자!
data1 = csv.reader(f)
header1=[]

next(data1)
header1.append(next(data1))
header1.append(next(data1))
data1=list(data1)
header1=list(header1)

for row in header1:
  print(row)

print()
# next(data)

for row in data1:
  for i in range(1, len(row)):
    if row[i] == '-':
      row[i]=0
    else:
      row[i]=float(row[i])
  print(row)
````

````
import pandas as pd

data1_d=pd.DataFrame(data1, columns=header1)
data1_d.set_index(data1_d[data1_d.columns[0]], inplace=True)
data1_d.index.name='종사자규모별'

data1_d.drop(data1_d.columns[0], axis=1, inplace=True)
data1_d=data1_d.transpose()


data1_d=data1_d.droplevel(axis=0, level=0)

data1_d=data1_d.loc[:,['1,000인 이상', '100인 이상~1,000인 미만', '10인 이상~100인 미만',	'10인 미만']]


data1_d
````
